# Prepare Migration from Article to Term

Let's get all articles (here: less than 100)

```python
endpoint = f"/integration/v1/article/"
r = requests.get(host+endpoint, headers=headers).json()
df = pd.DataFrame(r)
df
```

Prepare a dataframe with terms to upload

```python
terms = df.loc[:, ['title', 'body', 'custom_templates']]

terms['template_name'] = terms.custom_templates.apply(lambda t: t[0]['title'] if len(t) else '')
terms
```

```python
my_mapping = {
    "Fruit": {"template": 47, "glossary": 3},
    
}

my_list = []

for k, v in my_mapping.items():
  my_df = terms.loc[terms.template_name==k]
  for t in my_df.itertuples():
    my_list.append(dict(
        title = t.title,
        description = t.body,
        template_id = v['template'],
        glossary_ids = [v['glossary']]
    ))

my_list
```

# Post the Terms

Beware, there is no protection from creating duplicates at all.

```python
endpoint = f"/integration/v2/term/"
r = requests.post(host+endpoint, headers=headers, json=my_list).json()
r
```
