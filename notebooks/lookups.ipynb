```python
import requests
import pandas as pd
import io

countries = requests.get("https://davidmegginson.github.io/ourairports-data/countries.csv")
```

```python
df = pd.read_csv(io.StringIO(countries.text))
```

```python
df
```

Let's if any of the names or codes are duplicated. No entries is good, in this case

```python
df.loc[df.name.duplicated()]
```

```python
df.loc[df.code.duplicated()]
```

```python
df = df.set_index('code')
df
```

```python
df.loc['AG']
```

```python
name_to_code_mapping = df.loc[:, ['code']].set_index(df.name).squeeze()
name_to_code_mapping.head()
```

Let's try a different approach, creating a pandas Series from the start.

```python
name_to_code_mapping = pd.Series(data=list(df.code), index=df.name)
name_to_code_mapping
```

```python
name_to_code_mapping.get("Afghanistan")
```

```python
name_to_code_mapping.get("Afghanistaan", 'unknown')
```

```python
code_to_name_mapping = pd.Series(data=list(df.name), index=df.code)
code_to_name_mapping
```

```python
code_to_name_mapping['DE']
```
